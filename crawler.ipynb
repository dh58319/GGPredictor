```python
URL_skeleton = "https://sports.news.naver.com/kbaseball/news/index?date="

def URLs():
    Year = 2021
    Mon = 10
    date = 6
    while (Year == 2021):
        if date == 1:
            if Mon == 1:
                Mon = 12
                Year -= 1
            else: 
                Mon -= 1
                if Mon in [1,3,5,7,8,10,12]:
                    date = 31
                elif Mon == 2:
                    date = 28
                else:
                    date = 30
        else:
            date -= 1 
        URL = URL_skeleton + str(Year)+ str(Mon).zfill(2)+ str(date).zfill(2)
        time = str(Year)+ str(Mon).zfill(2)+ str(date).zfill(2)
        yield URL, time


for link, time in URLs():
    print(link, time)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
article_list = []
news = pd.DataFrame(columns=['title', 'Date'])


def article_Scraping():
    driver = webdriver.Chrome('./chromedriver')
    for link, date in URLs():
        print(link)
        for i in range(1,10):
            driver.get(link+"&page=" + str(i))
            time.sleep(0.5)
            articles = driver.find_elements(By.CSS_SELECTOR, '#_newsList > ul > li > div > a')
            for article in articles:
                if len(article.text) != 0:
                    news.loc[len(news)] = [article.text, date]       
    news.to_csv('./news2021.csv',sep=',',na_rep='NaN')
    driver.quit()
                
article_Scraping()
```
