```python
data = pd.read_csv('../crawl_data/news2022.csv')
selectdata = pd.DataFrame(data, columns=['title'])
selectdata
```

```python
pattern = r'\[[^)]*\\s\"\'\t]'
pattern2 = r'[^가-힣^a-z^A-Z^\.\:\s]'
for i in range(len(selectdata)):
    selectdata['title'][i]= re.sub(pattern, ' ',selectdata['title'][i])
    selectdata['title'][i]= re.sub(pattern2,'',selectdata['title'][i])
    print(selectdata['title'][i])
```

```python
with open('../stopword2.txt', 'r') as f:
    list_file = f.readlines()
stopwords =[]
for i in range(len(list_file)):
    stopwords.append(list_file[i][:-1])

#print(stopwords)
remove_stopwords= []
tokenized_doc = selectdata['title'].apply(lambda x: x.split()) # 토큰화
tokenized_doc = tokenized_doc.apply(lambda x: [item for item in x if item not in stopwords])
#for i in range(len(selectdata)):
    #if selectdata['title'][i] not in stopwords:
        #remove_stopwords.append(selectdata['title'][i])
```

```python
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
association_rules = apriori(tokenized_doc, min_support = 0.0005, min_confidence = 0.02, min_lift = 3, min_length=2)
association_result = list(association_rules);
```

```python
for item in association_result:
    #print(item[0])
    items = [x for x in item[0]]
    if '정후'  in items[0]:
        print(items[0] + " -> " + items[1])
        print("support = " + str(item[1]))
        print("confidence = " + str(item[2][0][2]))
        print("lift = " + str(item[2][0][3]))
        print()
```
