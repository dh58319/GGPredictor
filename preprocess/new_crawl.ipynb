```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
```

```python
article_list = []
news = pd.DataFrame(columns=['content' ,'Date'])
link ='https://sports.news.naver.com/kbaseball/news/index?date=20220819'
def article_Scraping():
    driver = webdriver.Chrome('./chromedriver')

    print(link)
    for i in range(1,10):
        driver.get(link+"&page=" + str(i))
        driver.implicitly_wait(0.5)
        articles = driver.find_elements(By.CSS_SELECTOR, '#_newsList > ul > li > div ')
        for article in articles:
            if len(article.text)!= 0:
                print(article.find_element(By.CSS_SELECTOR, 'a').text)


    news.to_csv('./news2022_full.csv',sep=',',na_rep='NaN')
    driver.quit()
```

```python
article_Scraping()
```
