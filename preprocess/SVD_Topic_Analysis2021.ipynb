```python
data = [['dummy']]
new_doc = pd.DataFrame(data, columns=['title'])
for i, document in enumerate(selectdata['title']):
    okt = Okt()
    clean_words = []
    for word in okt.pos(document, stem=True): #어간 추출
        if word[1] in ['Noun', 'Verb']: #명사, 동사
            clean_words.append(word[0])
    document = ' '.join(clean_words)
    new_doc.loc[i] = document
```

```python
with open('../stopword2.txt', 'r') as f:
    list_file = f.readlines()
stopwords =[]
for i in range(len(list_file)):
    stopwords.append(list_file[i][:-1])

print(stopwords)
remove_stopwords= []
tokenized_doc = new_doc['title'].apply(lambda x: x.split()) # 토큰화
tokenized_doc = tokenized_doc.apply(lambda x: [item for item in x if item not in stopwords])




```

```python
detokenized_doc = []
for i in range(len(selectdata)):

    t = ' '.join(tokenized_doc[i])
    detokenized_doc.append(t)

new_doc['new_title'] = detokenized_doc
```

```python
new_doc['new_title'] = new_doc['new_title'].apply(lambda x: ' '.join([w for w in x.split() if len(w)<5]))
new_doc
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
vectorizer = TfidfVectorizer(max_features= 10000, # 상위 1,000개의 단어를 보존
max_df = 0.5, smooth_idf=True)

X = vectorizer.fit_transform(new_doc['new_title'])

# TF-IDF 행렬의 크기 확인
print('TF-IDF 행렬의 크기 :',X.shape)
```

```python
svd_model = TruncatedSVD(n_components=40, algorithm='randomized', n_iter=100, random_state=122)
svd_model.fit(X)
len(svd_model.components_)
```

```python
terms = vectorizer.get_feature_names() # 단어 집합. 1,000개의 단어가 저장됨.

def get_topics(components, feature_names, n=5):
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx+1), [(feature_names[i], topic[i].round(5)) for i in topic.argsort()[:-n - 1:-1]])
get_topics(svd_model.components_,terms)

```
