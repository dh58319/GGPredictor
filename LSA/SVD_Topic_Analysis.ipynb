```python
selectdata['new_doc'] = selectdata['new_doc'].apply(lambda x: ' '.join([w for w in x.split() if len(w)<5]))
selectdata
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
```

```python
vectorizer = TfidfVectorizer(max_features= 1000, max_df = 0.5, smooth_idf=True)

X = vectorizer.fit_transform(selectdata['new_doc'])

# TF-IDF 행렬의 크기 확인
print('TF-IDF 행렬의 크기 :',X.shape)
```

```python
svd_model = TruncatedSVD(n_components=50, algorithm='randomized', n_iter=100, random_state=122)
svd_model.fit(X)
len(svd_model.components_)
```

```python
terms = vectorizer.get_feature_names() # 단어 집합. 1,000개의 단어가 저장됨.

def get_topics(components, feature_names, n=5):
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx+1), [(feature_names[i], topic[i].round(5)) for i in topic.argsort()[:-n - 1:-1]])
get_topics(svd_model.components_,terms)

```
